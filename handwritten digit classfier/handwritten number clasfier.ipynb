```python
import tensorflow as tf
from keras.datasets import mnist
```

```python
# load data
(x_train,y_train),(x_test,y_test)=mnist.load_data()
print(x_train.shape)
print(x_test.shape)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# visulization
fig, axes = plt.subplots(2,3)
fig.suptitle('handwritten digit')
for i in range(1,7):
    random_num=np.random.randint(0,len(x_train))
    plt.subplot(2,3,i)
    plt.imshow(x_train[random_num],cmap=plt.get_cmap('gray'))
    
plt.show()
    
```

```python
# preparing data set
img_rows=x_train[0].shape[0]
img_cols=x_train[0].shape[1]
print(img_rows,img_cols)
x_train=x_train.reshape(x_train.shape[0],img_rows,img_cols,1)
x_test=x_test.reshape(x_test.shape[0],img_rows,img_cols,1)
input_shape=(img_rows,img_cols,1)
x_train=x_train.astype('float32')
x_test=x_test.astype('float32')
x_train/=255
x_test/=255
print('x_train shape',x_train.shape)
```

```python
print(x_train)
```

```python
from keras.utils import np_utils
y_train=np_utils.to_categorical(y_train)
y_test=np_utils.to_categorical(y_test)
```

```python
print(y_train[0])
classes=y_train.shape[1]
print(y_train.shape[1])
```

# create model

```python

from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
from keras import backend as K
from keras.optimizers import SGD
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=input_shape))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(classes,activation='softmax'))

#compile
model.compile(loss='categorical_crossentropy',optimizer=SGD(0.01),metrics=['accuracy'])

```

```python
model.summary()
```

```python
batch_size=32
epochs=10
history=model.fit(x_train,y_train,batch_size=batch_size,epochs=epochs)
score=model.evaluate(x_test,y_test,verbose=1)
print('test loss',score[0])
print('test accuracy',score[1])
```

```python
#Loss plot
history_dict=history.history
loss_value=history_dict['loss']
epochs=range(1,11)
plt.title('Loss plot')
line=plt.plot(epochs,loss_value,label='train_loss')
plt.setp(line,linewidth=2.0,marker='*',markersize=10.0)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Accuracy plot
acc_value=history_dict['accuracy']
epochs=range(1,11)
line1=plt.plot(epochs,acc_value,label='train_accuracy')
plt.title('Accuracy plot')
plt.setp(line1,linewidth=2.0,marker='*',markersize=10.0)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Test loss and accuracy
print('Test loss',score[0])
print('Test accuracy',score[1])
model.save('model/hadwritten_digit_recognition_model.h5')
```
